Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
#Read in csv
chip_df = pd.read_csv('chipotle.csv')
```

```python
#Check
chip_df.head()
```

```python
#Recast item_price as a float
chip_df.item_price = chip_df.item_price.str.strip('$').astype('float')
```

```python
#Check
chip_df.info()
```

```python
#Check
chip_df.head()
```

```python
#Create series of four top grossing items
top_grossers = chip_df.groupby('item_name').agg('sum')['item_price'].nlargest(4)
```

```python
#Check
top_grossers
```

```python
#Plot the series
sns.barplot(x=top_grossers.index, y=top_grossers)
plt.xticks(rotation=45)
```

Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time

```python
#Look at sleepstudy documentation
data('sleepstudy', show_doc=True)
```

```python
#Read in sleepstudy dataset and check
ss = data('sleepstudy')
ss.head()
```

```python
#Format the subject column
ss['Subject'] = 'Subject_' + ss.Subject.astype(str)
```

```python
#Plot the results for each subject
plt.figure(figsize=(12, 8))
sns.lineplot(x='Days', y='Reaction', hue='Subject', data=ss)
```

```python
#Create a new dataframe of average reaction times by day
mean_ss = ss.groupby('Days').agg('mean')
mean_ss
```

```python
#Reset the index
mean_ss.reset_index(inplace=True)
mean_ss
```

```python
#Plot the new aggregate line on top of the original chart
plt.figure(figsize=(12, 8))
sns.lineplot(x='Days', y='Reaction', hue='Subject', data=ss)
sns.lineplot(x='Days', y='Reaction', color='black', linewidth='5', data=mean_ss)
```
